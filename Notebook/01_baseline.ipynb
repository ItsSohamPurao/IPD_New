```python
import numpy as np
import pandas as pd
from base_module import (
    get_preprocessed_data,
    train_and_evaluate,
    save_model_results,
    SVM_PARAMS
)

ALGORITHM_NAME = "Baseline"
DATA_PATH = r'D:\ACTUAL STUDY MATERIAL\IPD\Data\diabetic_data.csv'

print("="*70)
print(f"BASELINE MODEL (ALL FEATURES)")
print("="*70)
```

```python
print("\n--- Loading and Preprocessing Data ---")

X_train_encoded, X_test_encoded, y_train, y_test, \
    all_encoded_features, all_original_features, feature_mapping = \
    get_preprocessed_data(DATA_PATH)

print(f"\nBaseline uses ALL {len(all_encoded_features)} encoded features")
```

```python
print(f"\n--- Training Baseline Model ---")

clf, y_pred, report = train_and_evaluate(
    X_train_encoded, X_test_encoded,
    y_train, y_test,
    **SVM_PARAMS
)

print(f"\nBaseline Performance:")
print(f"  Accuracy: {report['accuracy']:.4f}")
print(f"  F1-score (class 1): {report['1']['f1-score']:.4f}")
```

```python
print(f"\n--- Saving Results ---")

# Baseline "selects" all features
save_model_results(
    algorithm_name=ALGORITHM_NAME,
    selected_encoded_features=all_encoded_features,
    all_encoded_features=all_encoded_features,
    all_original_features=all_original_features,
    feature_mapping=feature_mapping,
    report=report,
    results_csv='study_results.csv'
)

print("\n" + "="*70)
print("BASELINE COMPLETE")
print("="*70)
```
